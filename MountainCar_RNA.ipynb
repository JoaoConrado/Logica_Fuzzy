```python
!pip install pyvirtualdisplay
!apt-get install -y xvfb

# Importando as bibliotecas necessárias
from google.colab import drive
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from IPython.display import display, HTML
import numpy as np
import gym
import time
from gym.wrappers.record_video import RecordVideo
import io
import base64
from pyvirtualdisplay import Display
import glob

# Montando o Google Drive
drive.mount('/content/gdrive')

# Caminho para o arquivo CSV gerado pelo primeiro código
dados_csv = '/content/gdrive/MyDrive/Colab Notebooks/dftotal.csv'

# Carregando os dados
dados = pd.read_csv(dados_csv)

# Separando os previsores (Posição, Velocidade) e as classes (Ação)
previsores = dados[['Posicao', 'Velocidade']]
classe = dados['Acao']

# Dividindo os dados em treino e teste
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.17)

# Criando o modelo da RNA
classificador = Sequential()

# Primeira camada oculta
classificador.add(Dense(units=128, input_dim=2,
                        kernel_initializer='random_uniform',
                        activation='relu'))

# Segunda camada oculta
classificador.add(Dense(units=64,
                        kernel_initializer='random_uniform',
                        activation='relu'))

# Terceira camada oculta
classificador.add(Dense(units=32,
                        kernel_initializer='random_uniform',
                        activation='relu'))

# Camada de saída
classificador.add(Dense(units=3,  # Três ações possíveis (0, 1, 2)
                        kernel_initializer='random_uniform',
                        activation='softmax'))  # Softmax para multiclasse

# Compilando o modelo
classificador.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# Treinando a RNA
classificador.fit(previsores_treinamento, classe_treinamento,
                  batch_size=1, epochs=20)

# Configuração para gravação de vídeo
virtual_display = Display(visible=0, size=(1400, 900))
virtual_display.start()

def show_video():
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[-1]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        display(HTML(data=f'''<video alt="test" autoplay
                              controls style="height: 400px;">
                              <source src="data:video/mp4;base64,{encoded.decode('ascii')}" type="video/mp4" />
                           </video>'''))
    else:
        print("Não foi possível encontrar o vídeo.")

# Criando o ambiente do Mountain Car
env = RecordVideo(gym.make('MountainCar-v0'), './video')
obs = env.reset()

# Simulando o ambiente usando a RNA com lógica para reforçar o momentum
passo = 0
total_reward = 0

while True:
    passo += 1

    # Normalizando a observação para passar pela RNA
    entrada = np.array([obs]).reshape(1, -1)
    acao_probabilidades = classificador.predict(entrada, verbose=0)
    acao = np.argmax(acao_probabilidades)  # Seleciona a ação com maior probabilidade

    # Adicionando lógica para reforçar momentum
    if obs[1] < 0 and obs[0] < 0:  # Velocidade negativa e antes do topo esquerdo
        acao = 0  # Ação para acelerar para trás
    elif obs[1] > 0 and obs[0] > 0:  # Velocidade positiva e perto do topo direito
        acao = 2  # Ação para acelerar para frente

    obs, reward, terminate, _ = env.step(acao)
    total_reward += reward

    print(f"Passo: {passo}")
    print(f"Posição: {obs[0]}")
    print(f"Velocidade: {obs[1]}")
    print(f"Ação tomada: {acao}")
    print()

    # Condição para terminar a simulação após 500 passos
    if passo >= 500 or terminate:
        break

# Fechando o ambiente e mostrando o vídeo
env.close()
show_video()

print(f"Simulação finalizada em {passo} passos com uma recompensa total de {total_reward}.")
```
